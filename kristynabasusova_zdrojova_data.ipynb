```python
#import pandas - instalováno mám 
import pandas as pd
df_surovadata = pd.read_csv("kristynabasusova_surova_data.csv")
df_zdrojovadata = df_surovadata.copy()
```

```python
#kolik tam mám řádků a sloupců 
df_zdrojovadata.shape
```

```python
#chci vidět co jsem si načetla (prvních 5 řádků)

df_zdrojovadata.head()
```

```python
#datové typy 
df_zdrojovadata.info()
```

```python
# 1. Mohu přejmenovat plochu na plocha_v_m2 (protože všechno je m2) a následně vymazat sloupec plocha_jednotka

df_zdrojovadata["plocha_jednotka"].unique()
```

```python
df_zdrojovadata = df_zdrojovadata.rename(columns={"plocha": "plocha_v_m2"})
```

```python
df_zdrojovadata = df_zdrojovadata.rename(columns={"cena": "cena_v_czk"})
```

```python
df_zdrojovadata.head()
```

```python

# 2. Taky mohu změnit textová pole z object na string 

textove_sloupce = [
    "nazev",
    "ulice",
    "mesto",
    "dispozice",
    "mena",
    "perioda",
    "url"
]
df_zdrojovadata[textove_sloupce] = df_zdrojovadata[textove_sloupce].astype("string")
```

```python
#ještě přidám ID ke každému z bytů 

```

```python
#3. teď vymažu sloupec plocha_jednotka a předělám si pořadí 

df_zdrojovadata = df_zdrojovadata[['nazev', 'dispozice','plocha_v_m2','cena_v_czk','url','mesto','ulice']]
```

```python
df_zdrojovadata.head()
```

```python
#4. ceny jsou v tisích a bez desetinných míst, plocha také nemá desetinná místa, upravovat číslo asi není potřeba měnit typ na float
```

```python
#mohu ještě upravit první písmena sloupců na velké

df_zdrojovadata.columns = df_zdrojovadata.columns.str.capitalize()
```

```python
#exportuji upravená data 

df_zdrojovadata.to_csv('kristynabasusova_zdrojova_data.csv')
```
