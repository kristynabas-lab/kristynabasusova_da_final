```python
#3. dispozice bytu 

import re

odkazy_nabidek = soup.find_all('a', href=True)

for odkaz in odkazy_nabidek:
    href = odkaz.get('href', '')

    if '/detail/' not in href or '/byt/' not in href:
        continue

    dispozice = re.search(r"/byt/([^/]+)/", href)
    dispozice = dispozice.group(1) if dispozice else None

    if dispozice:
        print(dispozice)

```

```python
#2.rozměry bytu
import re

plochy = []

for nazev_nabidky in nazvy_nabidek:
    text = nazev_nabidky.text.strip()

    if not text.startswith("Pronájem bytu"):
        continue

    match = re.search(r"(\d+)\s*(m²|m2)", text)

    if match:
        plocha_hodnota = int(match.group(1))
        plocha_jednotka = match.group(2)

        plochy.append({
            "plocha": plocha_hodnota,
            "jednotka": plocha_jednotka
        })

plochy
```

```python
#5.lokace bytu (ulice + město)

ulice = []
mesta = []

for tag in nazvy_nabidek:
    text = tag.text.strip()

    # bereme jen adresy, ne názvy
    if text.startswith("Pronájem bytu"):
        continue

    parts = [p.strip() for p in text.split(",")]

    ulice.append(parts[0] if len(parts) > 0 else None)
    mesta.append(parts[1] if len(parts) > 1 else None)

ulice
mesta
```

```python
urls = []
dispozice_list = []

for a in soup.find_all('a', href=True):
    href = a.get('href', '')

    if '/detail/' not in href or '/byt/' not in href:
        continue

    url = "https://www.sreality.cz" + href
    urls.append(url)

    dispo = re.search(r"/byt/([^/]+)/", href)
    dispo = dispo.group(1) if dispo else None
    dispozice_list.append(dispo)
```

```python
# oprava/doplnění chyb - tady to je vyloženě konzultace s chatem

plocha_list = [p["plocha"] for p in plochy]
plocha_jednotka_list = [p["jednotka"] for p in plochy]
```

```python
nazev_list = []

for tag in nazvy_nabidek:
    text = tag.text.strip()
    if text.startswith("Pronájem bytu"):
        nazev_list.append(text)
```

```python
# kontrola 

n = min(
    len(nazev_list),
    len(urls),
    len(dispozice_list),
    len(ceny),
    len(meny),
    len(periody),
    len(plocha_list),
    len(plocha_jednotka_list),
    len(ulice),
    len(mesta),
)

print("Počet řádků do tabulky:", n)

nabidky = []

for i in range(n):
    nabidky.append({
        "nazev": nazev_list[i],
        "ulice": ulice[i],
        "mesto": mesta[i],
        "dispozice": dispozice_list[i],
        "plocha": plocha_list[i],
        "plocha_jednotka": plocha_jednotka_list[i],
        "cena": ceny[i],     
        "mena": meny[i],          
        "perioda": periody[i],    
        "url": urls[i],
    })

#vypsat první 2 nabídky abych to mohla zkontrolovat 
nabidky[:2]
```

```python
#import do csv 

import csv

with open("kristynabasusova_surova_data.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=nabidky[0].keys())
    writer.writeheader()
    writer.writerows(nabidky)

print("Hotovo: kristynabasusova_surova_data.csv")
```
